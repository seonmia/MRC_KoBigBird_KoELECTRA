## 예측 데이터_CSV와 답안 데이터_CSV 준비하기

'baseline.csv'파일을 업로드해준다.

```python
import pandas as pd
```

baseline.csv를 기준으로 10개의 데이터만 뽑아낸다.

```python
baseline = pd.read_csv('baseline.csv')
```

```python
train = baseline.head(10)
train
```

```python
target = train.copy()
```

```python
target
```

답안 데이터_CSV의 컬럼명을 Predicted => Answer로 바꿔준다.

```python
target.rename(columns={'Predicted': 'Answer'}, inplace=True)
target
```

Id, Predicted 컬럼을 가진 예측데이터와 Id, Answer 컬럼을 가진 답안 데이터를 만들었다.

혹여나 두 데이터 사이에 정렬된 순서가 다를수 있으니 Id 기준으로 병합한다.

```python
df_join = pd.merge(train, target, on='Id')
df_join
```

Predicted 컬럼과 Answer 컬럼사이에 편집거리함수를 적용해서 새로운 시리즈를 만든다.

```python
def edit_distance(row):
    s = row[1]
    t = row[2]
    m = len(s)+1
    n = len(t)+1
    D = [[0]*m for _ in range(n)]
    D[0][0] = 0
    
    for i in range(1,m):
        D[0][i] = D[0][i-1] + 1
    
    for j in range(1,n):
        D[j][0] = D[j-1][0] + 1
    
    for i in range(1,n):
        for j in range(1,m):
            cost = 0

            if s[j-1] != t[i-1]:
                cost = 1
            
            D[i][j] = min(D[i][j-1] + 1,D[i-1][j] + 1, D[i-1][j-1] + cost)
    
    return D[n-1][m-1]
```

```python
result = df_join.apply(edit_distance, axis = 1)
```

예측 데이터_CSV와 답안 데이터_CSV가 똑같아서 0이 나온다.

```python
result
```

편집거리의 평균이 최종 점수이다. 여기서는 당연히 0이다.

```python
result.sum()/len(result)
```

## 하나의 함수로 만들어 보자면

```python
from pprint import pprint
def evaluate_score(train_file, answer_file):
    """사용법: evaluate_score(예측파일_CSV의 경로명<문자열>, 답안파일_CSV의 경로명<문자열>)  
    단, 예측파일의 컬럼명은 Id, Predicted, 답안파일의 컬럼명은 Id, Answer로 세팅되어 있어야 한다.
    """
    train = pd.read_csv(train_file)
    target = pd.read_csv(answer_file)
    df_join = pd.merge(train, target, on='Id')
    def edit_distance(row):
        s = row['Predicted']
        t = row['Answer']
        m = len(s)+1
        n = len(t)+1
        D = [[0]*m for _ in range(n)]
        D[0][0] = 0
        
        for i in range(1,m):
            D[0][i] = D[0][i-1] + 1
        
        for j in range(1,n):
            D[j][0] = D[j-1][0] + 1
        
        for i in range(1,n):
            for j in range(1,m):
                cost = 0

                if s[j-1] != t[i-1]:
                    cost = 1
                
                D[i][j] = min(D[i][j-1] + 1,D[i-1][j] + 1, D[i-1][j-1] + cost)
        
        return D[n-1][m-1]
    result = df_join.apply(edit_distance, axis = 1)
    return result.sum()/len(result)
```

이전에 사용한 train 데이터 프레임과 target 데이터 프레임을 csv로 만들어서 테스트해본다

```python
train.to_csv('train.csv')
target.to_csv('target.csv')
```
